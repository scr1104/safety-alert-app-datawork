```python
import numpy as np
offencetype = ['Assault', 'Assault With Weapon', 'Robbery - Mugging', 'Robbery - Other', 'Robbery - Swarming',
              'Aggravated Assault', 'Discharge Firearm With Intent', 'Robbery - Purse Snatch', 'Pointing A Firearm', 
              'Theft From Mail / Bag / Key', 'Theft Over - Shoplifting', 'Assault - Force/Thrt/Impede', 
              'Robbery - Delivery Person', 'Use Firearm / Immit Commit Off', 'Theft Over - Distraction',
              'Theft - Misapprop Funds Over', 'Air Gun Or Pistol: Bodily Harm', 'Aggravated Assault Avails Pros',
              'Unlawfully Causing Bodily Harm']
df = MCI[MCI['offence'].isin(offencetype)]
df = df[df['premisetype'] == 'Outside']
```

```python
df[pd.isnull(df).any(axis=1)]
```

```python
df = df.dropna()
```

```python
df.columns
```

```python
df2 = df[['Long', 'Lat', 'offence', 'occurrencemonth', 'occurrenceyear', 'occurrencehour', 'occurrencedayofweek']]
df2.head()
```

```python
df2 = df2.reset_index()
df2 = df2.drop('X', axis=1)
```

```python
df2.head()
```

```python
len(offencetype)
```

```python
def month_num(x):
    month_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
                 'October', 'November', 'December']
    for i in range(1, 13):
        if x == month_list[i-1]:
            return i
df2['occurrencemonth'] = df2['occurrencemonth'].apply(month_num)
df2.head()
```

```python
def day_num(x):
    day_list = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
    for i in range(1, 8):
        if day_list[i-1] in x:
            return i
    else:
        return "NaN"
df3 = df2.copy()
df3['occurrencedayofweek_num'] = df3.occurrencedayofweek.apply(day_num)
df3.head()
```

```python
df4 = df3.copy()
df4.rename(columns = {"occurrencemonth":"month", 
                      "occurrenceyear":"year", "occurrencehour":"hour", 
                      "occurrencedayofweek":"dayofweek", "occurrencedayofweek_num":"dayofweek_num"})
```

```python
cols = df4.columns.tolist()
cols
```

```python
cols= cols[:3] + [cols[4], cols[3]] + cols[5:]
cols
```

```python
df = df4[cols] 
```

```python
df = df.sort_values(by=['occurrenceyear', 'occurrencemonth', 'occurrencehour'])
df = df.sort_index()
df.head()
```

```python
df.to_csv("Toronto_crime_data.csv")
pd.read_csv("Toronto_crime_data.csv")
```
